```python
import pandas as pd

# Load datasets
final_news_website = pd.read_csv('final_news_website.csv')
final_google_news = pd.read_csv('final_google_news.csv')
final_instagram_news= pd.read_csv('final_instagram_news.csv')

# Check the first few rows of the datasets to verify they are loaded correctly
print(final_news_website .head())
print(final_google_news .head())
print(final_instagram_news.head())
```

```python
# Check for missing values in each dataset
print("Missing values in final_news_website:")
print(final_news_website.isnull().sum())

print("Missing values in final_google_news:")
print(final_google_news.isnull().sum())

print("Missing values in final_instagram_news:")
print(final_instagram_news.isnull().sum())
```

```python
from sklearn.preprocessing import StandardScaler

# Initialize the scaler
scaler = StandardScaler()

# Apply scaling to the features (excluding the label)
features_columns = [f'feature_{i}' for i in range(20)]  # List of feature columns

# Scale the datasets
final_news_website[features_columns] = scaler.fit_transform(final_news_website[features_columns])
final_google_news[features_columns] = scaler.fit_transform(final_google_news[features_columns])
final_instagram_news[features_columns] = scaler.fit_transform(final_instagram_news[features_columns])

# Check the scaled data (optional)
print("Scaled final_news_website:")
print(final_news_website.head())

print("Scaled final_google_news:")
print(final_google_news.head())

print("Scaled final_instagram_news:")
print(final_instagram_news.head())
```

```python
from sklearn.model_selection import train_test_split

# Define feature columns and label column
features_columns = [f'feature_{i}' for i in range(20)]  # Features from feature_0 to feature_19
label_column = 'label'  # Label column

# Split the datasets into training and testing sets (80% training, 20% testing)
X_news_website = final_news_website[features_columns]
y_news_website = final_news_website[label_column]

X_google_news = final_google_news[features_columns]
y_google_news = final_google_news[label_column]

X_instagram_news = final_instagram_news[features_columns]
y_instagram_news = final_instagram_news[label_column]

# Split the data into 80% train and 20% test for each dataset
X_train_news_website, X_test_news_website, y_train_news_website, y_test_news_website = train_test_split(X_news_website, y_news_website, test_size=0.2, random_state=42)
X_train_google_news, X_test_google_news, y_train_google_news, y_test_google_news = train_test_split(X_google_news, y_google_news, test_size=0.2, random_state=42)
X_train_instagram_news, X_test_instagram_news, y_train_instagram_news, y_test_instagram_news = train_test_split(X_instagram_news, y_instagram_news, test_size=0.2, random_state=42)

# Check the shape of the splits
print("Training and Testing split for final_news_website:")
print(f"X_train: {X_train_news_website.shape}, X_test: {X_test_news_website.shape}")
print(f"y_train: {y_train_news_website.shape}, y_test: {y_test_news_website.shape}")

print("Training and Testing split for final_google_news:")
print(f"X_train: {X_train_google_news.shape}, X_test: {X_test_google_news.shape}")
print(f"y_train: {y_train_google_news.shape}, y_test: {y_test_google_news.shape}")

print("Training and Testing split for final_instagram_news:")
print(f"X_train: {X_train_instagram_news.shape}, X_test: {X_test_instagram_news.shape}")
print(f"y_train: {y_train_instagram_news.shape}, y_test: {y_test_instagram_news.shape}")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Initialize the model
model = LogisticRegression(max_iter=1000, random_state=42)

# Train the model on each dataset and evaluate
# For final_news_website
model.fit(X_train_news_website, y_train_news_website)
y_pred_news_website = model.predict(X_test_news_website)
print("Classification Report for final_news_website:")
print(classification_report(y_test_news_website, y_pred_news_website))
print("Confusion Matrix for final_news_website:")
print(confusion_matrix(y_test_news_website, y_pred_news_website))
print(f"Accuracy: {accuracy_score(y_test_news_website, y_pred_news_website)}\n")

# For final_google_news
model.fit(X_train_google_news, y_train_google_news)
y_pred_google_news = model.predict(X_test_google_news)
print("Classification Report for final_google_news:")
print(classification_report(y_test_google_news, y_pred_google_news))
print("Confusion Matrix for final_google_news:")
print(confusion_matrix(y_test_google_news, y_pred_google_news))
print(f"Accuracy: {accuracy_score(y_test_google_news, y_pred_google_news)}\n")

# For final_instagram_news
model.fit(X_train_instagram_news, y_train_instagram_news)
y_pred_instagram_news = model.predict(X_test_instagram_news)
print("Classification Report for final_instagram_news:")
print(classification_report(y_test_instagram_news, y_pred_instagram_news))
print("Confusion Matrix for final_instagram_news:")
print(confusion_matrix(y_test_instagram_news, y_pred_instagram_news))
print(f"Accuracy: {accuracy_score(y_test_instagram_news, y_pred_instagram_news)}\n")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming the datasets are loaded as final_news_website, final_google_news, final_instagram_news
# And that the target labels are in 'label' column

# Train-test split for final_news_website
X_train_news_website, X_test_news_website, y_train_news_website, y_test_news_website = train_test_split(final_news_website.drop('label', axis=1), final_news_website['label'], test_size=0.2, random_state=42)

# Train-test split for final_google_news
X_train_google_news, X_test_google_news, y_train_google_news, y_test_google_news = train_test_split(final_google_news.drop('label', axis=1), final_google_news['label'], test_size=0.2, random_state=42)

# Train-test split for final_instagram_news
X_train_instagram_news, X_test_instagram_news, y_train_instagram_news, y_test_instagram_news = train_test_split(final_instagram_news.drop('label', axis=1), final_instagram_news['label'], test_size=0.2, random_state=42)

# Model training for final_news_website
model_news_website = LogisticRegression(max_iter=1000)
model_news_website.fit(X_train_news_website, y_train_news_website)

# Model training for final_google_news
model_google_news = LogisticRegression(max_iter=1000)
model_google_news.fit(X_train_google_news, y_train_google_news)

# Model training for final_instagram_news
model_instagram_news = LogisticRegression(max_iter=1000)
model_instagram_news.fit(X_train_instagram_news, y_train_instagram_news)

# Evaluate News Website Model
y_pred_news_website = model_news_website.predict(X_test_news_website)
print("News Website Model Evaluation:")
print("Accuracy:", accuracy_score(y_test_news_website, y_pred_news_website))
print("Classification Report:\n", classification_report(y_test_news_website, y_pred_news_website))

# Confusion Matrix for News Website
cm_news_website = confusion_matrix(y_test_news_website, y_pred_news_website)
plt.figure(figsize=(6, 4))
sns.heatmap(cm_news_website, annot=True, fmt="d", cmap="Blues", xticklabels=['Class 0', 'Class 1'], yticklabels=['Class 0', 'Class 1'])
plt.title("Confusion Matrix - News Website Articles")
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()

# Evaluate Google News Model
y_pred_google_news = model_google_news.predict(X_test_google_news)
print("\nGoogle News Model Evaluation:")
print("Accuracy:", accuracy_score(y_test_google_news, y_pred_google_news))
print("Classification Report:\n", classification_report(y_test_google_news, y_pred_google_news))

# Confusion Matrix for Google News
cm_google_news = confusion_matrix(y_test_google_news, y_pred_google_news)
plt.figure(figsize=(6, 4))
sns.heatmap(cm_google_news, annot=True, fmt="d", cmap="Blues", xticklabels=['Class 0', 'Class 1'], yticklabels=['Class 0', 'Class 1'])
plt.title("Confusion Matrix - Google News Articles")
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()

# Evaluate Instagram News Model
y_pred_instagram_news = model_instagram_news.predict(X_test_instagram_news)
print("\nInstagram News Model Evaluation:")
print("Accuracy:", accuracy_score(y_test_instagram_news, y_pred_instagram_news))
print("Classification Report:\n", classification_report(y_test_instagram_news, y_pred_instagram_news))

# Confusion Matrix for Instagram News
cm_instagram_news = confusion_matrix(y_test_instagram_news, y_pred_instagram_news)
plt.figure(figsize=(6, 4))
sns.heatmap(cm_instagram_news, annot=True, fmt="d", cmap="Blues", xticklabels=['Class 0', 'Class 1'], yticklabels=['Class 0', 'Class 1'])
plt.title("Confusion Matrix - Instagram News Posts")
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Assuming the datasets have been loaded as final_news_website, final_google_news, final_instagram_news

# Split the datasets into features and labels
X_final_news_website = final_news_website.drop(columns=['label'])
y_final_news_website = final_news_website['label']

X_final_google_news = final_google_news.drop(columns=['label'])
y_final_google_news = final_google_news['label']

X_final_instagram_news = final_instagram_news.drop(columns=['label'])
y_final_instagram_news = final_instagram_news['label']

# Scale the features using StandardScaler
scaler = StandardScaler()

X_final_news_website_scaled = scaler.fit_transform(X_final_news_website)
X_final_google_news_scaled = scaler.fit_transform(X_final_google_news)
X_final_instagram_news_scaled = scaler.fit_transform(X_final_instagram_news)

# Train-Test split for final_news_website
X_train_news, X_test_news, y_train_news, y_test_news = train_test_split(X_final_news_website_scaled, y_final_news_website, test_size=0.2, random_state=42)
# Train-Test split for final_google_news
X_train_google, X_test_google, y_train_google, y_test_google = train_test_split(X_final_google_news_scaled, y_final_google_news, test_size=0.2, random_state=42)
# Train-Test split for final_instagram_news
X_train_instagram, X_test_instagram, y_train_instagram, y_test_instagram = train_test_split(X_final_instagram_news_scaled, y_final_instagram_news, test_size=0.2, random_state=42)

# Initialize the models
logreg = LogisticRegression(max_iter=1000)
nb = GaussianNB()
dt = DecisionTreeClassifier(random_state=42)
svc = SVC(kernel='linear', random_state=42)

# Define a function to train, evaluate and print results
def evaluate_model(model, X_train, X_test, y_train, y_test, model_name):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    print(f"{model_name} Classification Report:")
    print(classification_report(y_test, y_pred))
    
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Not Fake', 'Fake'], yticklabels=['Not Fake', 'Fake'])
    plt.title(f'Confusion Matrix for {model_name}')
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.show()
    
    accuracy = accuracy_score(y_test, y_pred)
    print(f"{model_name} Accuracy: {accuracy}\n")
    
    return accuracy

# Evaluate each model for final_news_website
print("Evaluating models for final_news_website:")
logreg_accuracy_news = evaluate_model(logreg, X_train_news, X_test_news, y_train_news, y_test_news, "Logistic Regression")
nb_accuracy_news = evaluate_model(nb, X_train_news, X_test_news, y_train_news, y_test_news, "Naive Bayes")
dt_accuracy_news = evaluate_model(dt, X_train_news, X_test_news, y_train_news, y_test_news, "Decision Tree")
svc_accuracy_news = evaluate_model(svc, X_train_news, X_test_news, y_train_news, y_test_news, "Linear SVC")

# Evaluate each model for final_google_news
print("Evaluating models for final_google_news:")
logreg_accuracy_google = evaluate_model(logreg, X_train_google, X_test_google, y_train_google, y_test_google, "Logistic Regression")
nb_accuracy_google = evaluate_model(nb, X_train_google, X_test_google, y_train_google, y_test_google, "Naive Bayes")
dt_accuracy_google = evaluate_model(dt, X_train_google, X_test_google, y_train_google, y_test_google, "Decision Tree")
svc_accuracy_google = evaluate_model(svc, X_train_google, X_test_google, y_train_google, y_test_google, "Linear SVC")

# Evaluate each model for final_instagram_news
print("Evaluating models for final_instagram_news:")
logreg_accuracy_instagram = evaluate_model(logreg, X_train_instagram, X_test_instagram, y_train_instagram, y_test_instagram, "Logistic Regression")
nb_accuracy_instagram = evaluate_model(nb, X_train_instagram, X_test_instagram, y_train_instagram, y_test_instagram, "Naive Bayes")
dt_accuracy_instagram = evaluate_model(dt, X_train_instagram, X_test_instagram, y_train_instagram, y_test_instagram, "Decision Tree")
svc_accuracy_instagram = evaluate_model(svc, X_train_instagram, X_test_instagram, y_train_instagram, y_test_instagram, "Linear SVC")

# Print accuracy summary
print("\nModel Comparison Summary:")
print(f"final_news_website - Logistic Regression Accuracy: {logreg_accuracy_news}")
print(f"final_news_website - Naive Bayes Accuracy: {nb_accuracy_news}")
print(f"final_news_website - Decision Tree Accuracy: {dt_accuracy_news}")
print(f"final_news_website - Linear SVC Accuracy: {svc_accuracy_news}")

print(f"final_google_news - Logistic Regression Accuracy: {logreg_accuracy_google}")
print(f"final_google_news - Naive Bayes Accuracy: {nb_accuracy_google}")
print(f"final_google_news - Decision Tree Accuracy: {dt_accuracy_google}")
print(f"final_google_news - Linear SVC Accuracy: {svc_accuracy_google}")

print(f"final_instagram_news - Logistic Regression Accuracy: {logreg_accuracy_instagram}")
print(f"final_instagram_news - Naive Bayes Accuracy: {nb_accuracy_instagram}")
print(f"final_instagram_news - Decision Tree Accuracy: {dt_accuracy_instagram}")
print(f"final_instagram_news - Linear SVC Accuracy: {svc_accuracy_instagram}")
```
